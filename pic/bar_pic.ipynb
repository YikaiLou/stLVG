```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
methods = ['SLAT', 'BANKSY', 'stLVG']
merfish_data = [0.5346, 0.6196, 0.6083]
starmap_data = [0.6939, 0.7152, 0.8557]
barista_data = [0.6481, 0.6062, 0.7635]

# 设置图形和子图
fig, axs = plt.subplots(1, 3, figsize=(10, 5))

# 绘制第一个子图
axs[0].bar(methods, merfish_data, color=["#9DD79D", "#56B3E8", "#FDB815"], width = 0.7, edgecolor='black', linewidth=1)
# axs[0].set_title('MERFISH', fontsize = 36)
# axs[0].set_ylabel('ARI', fontsize = 36, fontweight='bold')
axs[0].tick_params(axis='both', which='major', labelsize=22, direction='in')
axs[0].set_ylim(0, 1)
axs[0].tick_params(axis='y', which='major', labelsize=28)
axs[0].set_xticklabels([])
# 绘制第二个子图
axs[1].bar(methods, starmap_data, color=["#9DD79D", "#56B3E8", "#FDB815"], width = 0.7, edgecolor='black', linewidth=1)
# axs[1].set_title('STARmap', fontsize = 36)
axs[1].tick_params(axis='both', which='major', labelsize=12, direction='in')
axs[1].set_ylim(0, 1)
axs[1].set_xticklabels([])
axs[1].set_yticklabels([])
# 绘制第三个子图
axs[2].bar(methods, barista_data, color=["#9DD79D", "#56B3E8", "#FDB815"], width = 0.7, edgecolor='black', linewidth=1)
# axs[2].set_title('BaristaSeq', fontsize = 36)
axs[2].tick_params(axis='both', which='major', labelsize=12, direction='in')
axs[2].set_ylim(0, 1)
axs[2].set_xticklabels([])
axs[2].set_yticklabels([])
# 显示图形
plt.tight_layout()
plt.show()
```

```python
#BAS1

ind = ['ARI', 'NMI']
slat_v = [0.7858, 0.8211]
banksy_v = [0.6947, 0.8194]
stLVG_v = [0.8011, 0.8243]
x = np.arange(len(ind))
width = 0.2

fig, ax = plt.subplots(figsize=(5, 4))

ax.bar(x - width, slat_v, width, alpha=0.9, color="#c2c2f0", label='SLAT')
ax.bar(x, banksy_v, width, alpha=0.9, color="#66b3ff", label='BANKSY')
ax.bar(x + width, stLVG_v, width, alpha=0.9, color='#ff9999', label='stLVG')

ax.set_xlabel('')
ax.set_ylabel('')
ax.set_title('')
ax.xaxis.set_ticklabels([])
ax.xaxis.set_ticks([])
# ax.yaxis.set_ticklabels([])
# ax.yaxis.set_ticks([])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.show()
```
